```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt

df = pd.read_csv("student_data (1).csv")

X = df["StudyHours"].values.reshape(-1,1)
y = df["ExamScore"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=3)
model = LinearRegression()
model.fit(X_train,y_train)

plt.scatter(df["StudyHours"],df["ExamScore"],color='red',label="Actual Data")
plt.plot(X_train,model.predict(X_train),color='blue',label='Predicted Values')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")

mse = mean_squared_error(y_test,model.predict(X_test))
r2 = r2_score(y_test,model.predict(X_test))

print(r2)
print(mse)
```
