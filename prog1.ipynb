```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.stats import t

data = {
    "plantname" :["plant 1","plant 2","plant 3","plant 4"],
    "sun" :[3,4,5,6],
    "height" :[1,5,3,4]
}

df=pd.DataFrame(data)
x=df["sun"]
y=df["height"]
xmean=np.mean(x)
ymean=np.mean(y)

plt.scatter(df["height"],df["sun"],color='blue')
plt.title("plant height vs Sun Exposure")
plt.xlabel("Sun Light")
plt.ylabel("plant height")

numerator = np.sum((x-xmean)*(y-ymean))
denominator = np.sqrt(sum((x-xmean)**2)*sum((y-ymean)**2))
corr = numerator/denominator
print(corr)

n=len(x)
t_statistic = corr * np.sqrt(n-1 // (1-corr**2))
p_value = 1-t.cdf(abs(t_statistic),df=n-1)
print("t statistic value is : ",t_statistic)
print("p value is : ",p_value)

if p_value < 0.05 :
    print("There is statistically significant relation ship between sunlight and plant height")
else : print("There is no statistically significant relationship between plant height and sunlight")

if corr > 0 :
    print("There is Positive Correlation between plant height and sunlight")
elif corr < 0 :
    print("There is negative correlation between plant height and sunlight")
else : print ("There is no Correlation between plant height and sunlight")
```
